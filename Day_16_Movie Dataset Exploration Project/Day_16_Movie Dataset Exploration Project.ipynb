```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("movie.csv")
df.head()
```

```python
# Display basic information about the dataset
print("Dataset Info:")
print(df.info())

# Display summary statistics
print("\nSummary Statistics:")
print(df.describe())
```

```python
# Count males who watched Barbie
male_barbie_count = df[(df['is_male'] == True) & (df['movie'] == 'Barbie')].shape[0]

print(f"The number of males who watched the Barbie movie is: {male_barbie_count}")
```

```python
# Insight 1: Average Age by Gender
avg_age_by_gender = df.groupby('is_male')['age'].mean()
print("Average Age by Gender:")
print(avg_age_by_gender)
```

```python
# Insight 2: Most Watched Movie
most_watched_movie = df['movie'].value_counts().idxmax()
print("\nMost Watched Movie:")
print(most_watched_movie)
```

```python
# Insight 3: Gender Distribution
gender_distribution = df['is_male'].value_counts(normalize=True) * 100
print("\nGender Distribution:")
print(gender_distribution)
```

```python
# Insight 4: Age Distribution by Movie
plt.figure(figsize=(10, 6))
sns.boxplot(x='movie', y='age', data=df, palette='pastel')
plt.title('Age Distribution by Movie')
plt.xlabel('Movie')
plt.ylabel('Age')
plt.show()
```

```python
# Insight 5: Movie Preference by Gender
movie_preference_by_gender = df.groupby(['is_male', 'movie']).size().unstack()
print("\nMovie Preference by Gender:")
print(movie_preference_by_gender)
```

```python
# Insight 6: Correlation Matrix
correlation_matrix = df.corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=.5)
plt.title('Correlation Matrix')
plt.show()
```

```python
# Visualize the distribution of 'age'
plt.figure(figsize=(10, 6))
sns.histplot(df['age'], bins=20, kde=True, color='skyblue')
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# Countplot for movie genres
plt.figure(figsize=(8, 5))
sns.countplot(x='movie', data=df, hue='is_male', palette='viridis')
plt.title('Count of Movies by Gender')
plt.xlabel('Movie')
plt.ylabel('Count')
plt.show()
```

```python
# Pairplot to explore relationships
sns.pairplot(df, hue='movie', palette='husl')
plt.suptitle('Pairplot of Age, Gender, and Movie')
plt.show()
```
