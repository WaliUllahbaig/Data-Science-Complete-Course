```python
# On Image
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image
image = cv2.imread("img.png", cv2.IMREAD_GRAYSCALE)  # Read the image in grayscale

# Display the original image
plt.subplot(121), plt.imshow(image, cmap='gray'), plt.title('Original Image')

#filter for horizontal edges
# Define a 3x3 filter (e.g., a simple averaging filter)
filter_array = np.array([[0, 1, 0],
                         [0, -1, 0],
                         [0, 0, 0]])

# Apply the filter using convolution
filtered_image = cv2.filter2D(image, -1, filter_array)

# Display the filtered image
plt.subplot(122), plt.imshow(filtered_image, cmap='gray'), plt.title('Filtered Image')
plt.show()
```

```python
#filter for verticle edges
# Define a 3x3 filter (e.g., a simple averaging filter)
filter_array = np.array([[0, 0, 0],
                         [0, -1, 1],
                         [0, 0, 0]])

# Apply the filter using convolution
filtered_image = cv2.filter2D(image, -1, filter_array)

# Display the filtered image
plt.subplot(122), plt.imshow(filtered_image, cmap='gray'), plt.title('Filtered Image')
plt.show()
```

```python
#Prewit Filter

# Define a 3x3 filter (e.g., a simple averaging filter)
filter_array = np.array([[1, 0, -1],
                         [1, 0, -1],
                         [1, 0, -1]])
# Apply the filter using convolution
filtered_image = cv2.filter2D(image, -1, filter_array)

# Display the filtered image
plt.subplot(122), plt.imshow(filtered_image, cmap='gray'), plt.title('Filtered Image')
plt.show()
```

```python
#Sobel Filter

# Define a 3x3 filter (e.g., a simple averaging filter)
filter_array = np.array([[1, 0, -1],
                         [2, 0, -2],
                         [1, 0, -1]])
# Apply the filter using convolution
filtered_image = cv2.filter2D(image, -1, filter_array)

# Display the filtered image
plt.subplot(122), plt.imshow(filtered_image, cmap='gray'), plt.title('Filtered Image')
plt.show()
```

```python
#Roberts Filter

# Define a 3x3 filter (e.g., a simple averaging filter)
filter_array = np.array([[0, 0, -1],
                         [0, 1, 0],
                         [0, 0, 0]])
# Apply the filter using convolution
filtered_image = cv2.filter2D(image, -1, filter_array)

# Display the filtered image
plt.subplot(122), plt.imshow(filtered_image, cmap='gray'), plt.title('Filtered Image')
plt.show()
```

```python
#Verticle Filter

# Define a 3x3 filter (e.g., a simple averaging filter)
filter_array = np.array([[1, 1, 1],
                         [0, 0, 0],
                         [-1, -1, -1]])
# Apply the filter using convolution
filtered_image = cv2.filter2D(image, -1, filter_array)

# Display the filtered image
plt.subplot(122), plt.imshow(filtered_image, cmap='gray'), plt.title('Filtered Image')
plt.show()
```
